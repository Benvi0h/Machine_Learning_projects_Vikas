```python
import pandas as pd
import numpy as np
import warnings 
import os 

warnings.filterwarnings('ignore')
```

```python
with open('Family_data.json') as file:
    data = json.load(file)
```

```python
len(data)
```

```python
questions = [i['question'] for i in data[:32]]
answers = [i['answer'] for i in data[:32]]
```

```python
new_data = []
for q, a in zip(questions, answers):
    new_data.append({'questions' : q, 'answers' : a})
new_data_df = pd.DataFrame(new_data)
```

```python
k_df = pd.DataFrame(np.array([questions, answers]).transpose(), columns=['questions', 'answers'])
```

```python
batch_size = 32

for batch in range(len(data) // batch_size):
    # current_batch = data[batch_size * batch : batch_size * (batch + 1)]
    if batch < len(data) // batch_size - 1:
        print(f"{batch_size * batch} : {batch_size * (batch + 1)}")
        current_batch = data[batch_size * batch : batch_size * (batch + 1)]
        
    else:
        print(f"{batch_size * batch} : ")
        current_batch = data[batch_size * batch :]
        
    # for pair in current_batch:
    
    questions = [i['question'] for i in current_batch]
    answers = [i['answer'] for i in current_batch]

    new_data = []
    for q, a in zip(questions, answers):
        new_data.append({'questions' : q, 'answers' : a})
    new_data_df = pd.DataFrame(new_data)

    if os.path.exists('Family_data.csv'):
        temp_df = pd.read_csv('Family_data.csv')
        temp_df = pd.concat([temp_df, new_data_df], ignore_index=True)
        temp_df.to_csv('Family_data.csv', index=False)
    else:
        temp_df = pd.DataFrame(new_data, columns=['questions', 'answers'])
        temp_df.to_csv('Family_data.csv', index=False)    
```
