```python
import numpy as np
import matplotlib.pyplot as plt
# import seaborn
import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc
from scipy.ndimage import uniform_filter1d
import seaborn as sns

import tensorflow as tf
import tensorflow
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Layer, Dropout, Dense, ReLU, Activation
from tensorflow.keras.activations import sigmoid

import warnings
warnings.filterwarnings("ignore")
```

```python
# Load the datasets
dataset = [None]*10
dataset[0] = pd.read_csv("linear_sudden_rotation_noise_and_redunce.csv")

column_names_hyper = ["x1","x2","x3","x4","x5","x6","x7","x8","x9","x10"]
hyper_df = pd.read_csv("Rotating_hyperplane/rotatingHyperplane.data", delimiter=" ", header=None, names = column_names_hyper)
hyper_lables = pd.read_csv("Rotating_hyperplane/rotatingHyperplane.labels", delimiter=" ", header=None, names = ["label"])
dataset[1] = pd.concat([hyper_df, hyper_lables], axis = 1)

column_names_sea = ["x1","x2","x3"]
sea_df = pd.read_csv("Sea_training/SEA_training_data.csv", header = None, names = column_names_sea)
sea_labels = pd.read_csv("Sea_training/SEA_training_class.csv", header = None, names=["label"])
dataset[2] = pd.concat([sea_df, sea_labels], axis = 1)

column_names_stagger = ["x1","x2", "x3","label"]
dataset[3] = pd.read_csv("harvard_dataverse_files/sea_0123_abrupto_noise_0.2.csv")
dataset[3].columns = column_names_stagger
dataset[4] = pd.read_csv("harvard_dataverse_files/sea_0123_gradual_noise_0.2.csv")
dataset[4].columns = column_names_stagger
```

```python
class BNN(Model):
    def __init__(self, input_size, hidden_size, output_size, dropout_rate=0.2):
        super(BNN, self).__init__()
        self.fc1 = Dense(hidden_size, input_shape=(input_size,))
        self.dropout = Dropout(dropout_rate)
        self.fc2 = Dense(output_size)
        self.relu = ReLU()
        self.sigmoid = Activation('sigmoid')
        
    def call(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.dropout(x)
        x = self.fc2(x)
        x = self.sigmoid(x)
        return x

def monte_carlo_predictions(model, X, n_samples=50):
    # model.train()  # Enable dropout
    preds = tf.stack([model(X, training=True) for _ in range(n_samples)], axis=0)
    mean_pred = tf.reduce_mean(preds, axis=0)
    uncertainty = tf.math.reduce_variance(preds, axis=0)
    return mean_pred, uncertainty
    
def detect_concept_drift(uncertainty, threshold):
    return tf.reduce_mean(uncertainty).numpy() > threshold

def train(model, loss_fn, optimizer, X_train, y_train, epochs=100):
    for epoch in range(epochs):
        with tf.GradientTape() as tape:
            outputs = model(X_train, training=True)  # Forward pass with training=True
            loss = loss_fn(y_train, outputs)        # Compute the loss

        # Compute gradients and update weights
        gradients = tape.gradient(loss, model.trainable_variables)
        optimizer.apply_gradients(zip(gradients, model.trainable_variables))

def evaluate_metrics(y_true, y_pred):
    # Compute confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    print("Confusion Matrix:")
    print(cm)

    # Print classification report
    report = classification_report(y_true, y_pred, target_names=["Class 0", "Class 1"])
    print("\nClassification Report:")
    print(report)

    return cm, report

def plot_roc(y_true, y_pred_prob, name_of_file):
    fpr, tpr, _ = roc_curve(y_true, y_pred_prob)
    roc_auc = auc(fpr, tpr)
    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("Receiver Operating Characteristic (ROC) Curve")
    plt.legend(loc="lower right")
    plt.savefig(f"results_images/{name_of_file}.png", dpi = 300)
    plt.show()

def evaluate_drift_detection(batch_size, number_of_samples, drift_point, drift_points, error_difference = 50):
    true_positives = []
    total_drift_points = 0
    
    drift_batch  = round(drift_point / batch_size)
    number_of_batches = int (round(number_of_samples / batch_size))
    # detected = 0
    
    drift_batches_to_be_detected = []
    for i in range(number_of_batches):
        if(i % drift_batch == 0):
            drift_batches_to_be_detected.append(i)

    
    for point in drift_points:
        # print(batch)
        for batch in drift_batches_to_be_detected:
            # if point % drift_batch >=0 and point % drift_batch < error_difference and point > drift_batch:
            
            if abs(point - batch) < error_difference:
                print(f"{point} - {batch} = {abs(point - batch)}")
                true_positives.append(point)
                break
    
    
    n_true_positives = len(true_positives)
    
    
    
    # number_of_batches

    
    # for batch in drift_batches_to_be_detected:
    #     for pos in true_positives:
    #         # print(batch)
    #         if abs(pos - batch) < error_difference:
    #             # print(f"{pos} - {batch} = {abs(pos - batch)}")
    #             detected += 1
    #             break
    n_false_positives = len(drift_points) - n_true_positives
    misses = len(drift_batches_to_be_detected) - n_true_positives
    print(f"drift batch : {drift_batch}")
    print(f"Total number of samples : {number_of_samples}")
    print(f"Batches to be detected : {len(drift_batches_to_be_detected)}")
    print(f"Drift point : {drift_point}")
    print(f"batch_size : {batch_size}")
    print(f"Number of batches : {number_of_batches}")
    print(f"Number of detected drift points : {len(drift_points)}")
    print(f"Number of true positives : {n_true_positives}")
    print(f"Number of false positives : {n_false_positives}")
    print(f"Number of misses : {misses}")
```

# linear_sudden_rotation_noise_and_redunce dataset
### Settings
- batch_size = 1000
This is done due to the fact that there are drift points every 1000 sample records. This way recording the models performance and its abilty to detect drift can be evaluated easily

### Summary
- The number of record in this is dataset : 100000
- learning_rate : 0.1
- dropout : 0.2
- uncertainity threshold (variable name : threshold) : 0.05
- model performance : 1.0
- number of epochs : 50

```python
## Change the dataset as required
data = dataset[0] 
columns = data.columns.tolist()[:-1]
# Extract features and labels
X_data = tf.convert_to_tensor(data[columns].values, dtype=tf.float32)
y_data = tf.convert_to_tensor(data["label"].values, dtype=tf.float32)
y_data = tf.reshape(y_data, [-1, 1])  # Reshape to column vector

# Simulate streaming batches
batch_size = 1000
def create_batches(X, y, batch_size):
    return [(X[i:i+batch_size], y[i:i+batch_size]) for i in range(0, len(X), batch_size)]

batches = create_batches(X_data, y_data, batch_size)
```

```python
len(batches)
```

```python
# Initialize the model, criterion, and optimizer
input_size = 5
hidden_size = 15
output_size = 1
model = BNN(input_size, hidden_size, output_size, dropout_rate=0.1)
criterion = tf.keras.losses.BinaryCrossentropy()
optimizer = tf.keras.optimizers.Adam(learning_rate=0.01)

# Train on the first batch
X_batch, y_batch = batches[0]
train(model, criterion, optimizer, X_batch, y_batch, epochs=150)

# Track uncertainty, predictions, and labels
uncertainties = []
accuracies = []
drift_points = []
batch_indices = []

all_preds = []
all_probs = []
all_true = []
thresholds = []

# Process subsequent batches
number_of_batches = len(batches) 
for i, (X_batch, y_batch) in enumerate(batches[0:], start=1):
    print(f"\rbatch {i} of {number_of_batches}", end = "")
    # Monte Carlo predictions
    mean_pred, uncertainty = monte_carlo_predictions(model, X_batch)
    uncertainties.append(tf.reduce_mean(uncertainty).numpy())
    
    # Calculate dynamic threshold
    if i > 1:
        smoothed_uncertainties = uniform_filter1d(uncertainties, size=3)
        threshold = np.mean(smoothed_uncertainties) + np.std(smoothed_uncertainties)
    else:
        threshold = 0.05 # Default threshold for the first batch
    
    thresholds.append(threshold)
    # Detect drift
    if detect_concept_drift(uncertainty, threshold):
        
        print(f"\nDrift detected at batch {i} of {number_of_batches}. Retraining...")
        train(model, criterion, optimizer, X_batch, y_batch, epochs=100)
        drift_points.append(i)
```

```python
all_preds = []
all_probs = []
all_true = []
accuracies = []
batch_indices = []

# No need for model.eval() in TensorFlow; dropout is controlled via `training` flag.

mean_pred = tf.convert_to_tensor(mean_pred)  # Ensure mean_pred is a TensorFlow tensor
y_batch = tf.convert_to_tensor(y_batch)      # Ensure y_batch is a TensorFlow tensor

# Prediction probabilities
preds_prob = mean_pred.numpy().flatten()  # Convert TensorFlow tensor to NumPy array and flatten

# Binary predictions
preds = tf.cast(mean_pred > 0.5, tf.float32).numpy().flatten()  # Convert to NumPy

# True labels
y_true = y_batch.numpy().flatten()  # Convert TensorFlow tensor to NumPy array and flatten

# Store predictions and true labels for final evaluation
all_preds.extend(preds)
all_probs.extend(preds_prob)
all_true.extend(y_true)

# Compute batch-wise accuracy
accuracy = np.mean(preds == y_true)  # NumPy mean for comparison
accuracies.append(accuracy)
batch_indices.append(i)
```

```python
# Final evaluation
print("\nFinal Evaluation:")
cm, report = evaluate_metrics(all_true, all_preds)
plot_roc(all_true, all_probs, name_of_file=f"roc_curve_linear_rotation_sudden_dataset_{batch_size}")
```

```python
# Confusion matrix heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=["Class 0", "Class 1"], yticklabels=["Class 0", "Class 1"])
plt.title("Final Confusion Matrix")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.savefig(f"results_images/confusion_matrix_linear_rotation_sudden_dataset_{batch_size}.png", dpi = 300)
plt.show() 
```

```python
# Accuracy over batches
plt.figure(figsize=(17, 5))
plt.subplot(1, 2, 1)
plt.plot(batch_indices, accuracies, label="Accuracy")
for point in drift_points:
    plt.axvline(x=point, color='red', linestyle='--', label='Drift Point' if point == drift_points[0] else "")
plt.xlabel("Batch Index")
plt.ylabel("Accuracy")
plt.title("Model Accuracy Over Batches")
plt.legend()
```

```python
# Initialize the model, criterion, and optimizer
input_size = 5
hidden_size = 15
output_size = 1
model = BNN(input_size, hidden_size, output_size, dropout_rate=0.3)
criterion = tf.keras.losses.BinaryCrossentropy()
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)

# Train on the first batch
X_batch, y_batch = batches[0]
train(model, criterion, optimizer, X_batch, y_batch, epochs=200)

# Track uncertainty, predictions, and labels
uncertainties = []
accuracies = []
drift_points = []
batch_indices = []

all_preds = []
all_probs = []
all_true = []

thresholds = []

# Process subsequent batches
number_of_batches = len(batches) 
for i, (X_batch, y_batch) in enumerate(batches[0:], start=1):
    print(f"\rbatch {i} of {number_of_batches}", end = "")
    # Monte Carlo predictions
    mean_pred, uncertainty = monte_carlo_predictions(model, X_batch)
    uncertainties.append(tf.reduce_mean(uncertainty).numpy())
    
    # Calculate dynamic threshold
    if i > 1:
        smoothed_uncertainties = uniform_filter1d(uncertainties, size=3)
        threshold = np.mean(smoothed_uncertainties) + np.std(smoothed_uncertainties)
    else:
        threshold = 0.05 # Default threshold for the first batch
    
    thresholds.append(threshold)
    # Detect drift
    if detect_concept_drift(uncertainty, threshold):
        
        print(f"\nDrift detected at batch {i} of {number_of_batches}. Retraining...")
        train(model, criterion, optimizer, X_batch, y_batch, epochs=150)
        drift_points.append(i)
```

```python
all_preds = []
all_probs = []
all_true = []
accuracies = []
batch_indices = []

# No need for model.eval() in TensorFlow; dropout is controlled via `training` flag.

mean_pred = tf.convert_to_tensor(mean_pred)  # Ensure mean_pred is a TensorFlow tensor
y_batch = tf.convert_to_tensor(y_batch)      # Ensure y_batch is a TensorFlow tensor

# Prediction probabilities
preds_prob = mean_pred.numpy().flatten()  # Convert TensorFlow tensor to NumPy array and flatten

# Binary predictions
preds = tf.cast(mean_pred > 0.5, tf.float32).numpy().flatten()  # Convert to NumPy

# True labels
y_true = y_batch.numpy().flatten()  # Convert TensorFlow tensor to NumPy array and flatten

# Store predictions and true labels for final evaluation
all_preds.extend(preds)
all_probs.extend(preds_prob)
all_true.extend(y_true)

# Compute batch-wise accuracy
accuracy = np.mean(preds == y_true)  # NumPy mean for comparison
accuracies.append(accuracy)
batch_indices.append(i)
```
